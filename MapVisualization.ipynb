```python
import folium
import pandas as pd
from folium import plugins

m=folium.Map(location=[40.70186402, -74.01590257],width=500,height=500,zoom_start=12)
```

```python

data = pd.read_csv('df_naive.csv') 


LOS = folium.map.FeatureGroup()
for lat, lng, in zip(data.lat, data.lon):
    LOS.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )
m.add_child(LOS)

```

```python
from folium import plugins

m=folium.Map(location=[40.70186402, -74.01590257],width=500,height=500,zoom_start=12)

LOS = plugins.MarkerCluster().add_to(m)

for lat, lng in zip(data.lat, data.lon):
    folium.Marker(
        location=[lat, lng],
        icon=None,
    ).add_to(LOS)

m.add_child(LOS)
```

```python
from folium.plugins import HeatMap

m=folium.Map(location=[40.71386402, -73.95590257],width=500,height=500,zoom_start=10.3)

# Convert data format
heatdata = data[['lat','lon']].values.tolist()
# add incidents to map
HeatMap(heatdata).add_to(m)
m
```
